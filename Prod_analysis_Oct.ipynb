```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Premalatha-success/Datasets/refs/heads/main/CardioGoodFitness-1.csv')
display(df.head())
```

```python
product_counts = df['Product'].value_counts()
display(product_counts)
```

```python
product_gender_counts = pd.crosstab(df['Product'], df['Gender'])
display(product_gender_counts)
```

```python
product_marital_counts = pd.crosstab(df['Product'], df['MaritalStatus'])
display(product_marital_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.boxplot(x='Product', y='Age', data=df)
plt.title('Distribution of Age by Product')
plt.xlabel('Product')
plt.ylabel('Age')
plt.show()
```
